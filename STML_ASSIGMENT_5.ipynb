```
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler, LabelEncoder

url_salary = '/content/Salary_Data.csv'
df_salary = pd.read_csv(url_salary)

features_salary = df_salary[['YearsExperience']]
target_salary = df_salary['Salary']


plt.figure(figsize=(8, 6))
plt.scatter(features_salary, target_salary, color='blue')
plt.title('Years of Experience vs Salary')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()


scaler = MinMaxScaler()
normalized_salary = scaler.fit_transform(df_salary)
```
